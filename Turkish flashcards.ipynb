# Now we get to the flashcards!

- Choose a category.
- An English word from that category is printed.
- Enter what you think the Turkish translation is.
    - If it is correct, it will be removed from the dictionary.
    - Score: every correct answer adds 1, every incorrect answer subtracts 1.
- Continue until the dictionary is empty.
- When you finish you can choose another category or end your lesson.

* If at any point you want to quit, simply enter '0' into the user input box.

<Work in progress: move entries to review dictionary to review later rather than deleting them>

```python
from sys import exit
import random

dct = {}

def success():
    print("""
^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
Good work today!!!! 
^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^""")
    print(""" 
                       _____
                      |    ||
                      |    ||
                      |      \\
                       \\      \\ 
                        \\      \\
       ---------------- v       \\
      / '         \\              \\
      \\___________/     \\         \\
     /  '         \\      \\         \\
     \____________/                 \\
     / '          \                 |
     \____________/                 |
      / '        \                 /
      \__________/                /
       --------------------------/
       """)
    again = input("Press 1 to choose another category or 0 to exit.")
    if '1' in again:
        category()
    else:
        fail()
        
def fail():
    print("""
    See you next time!""")
    print("""
                          ___     
              \     ___  |   |  ___
           \   \   |   | |   | |   |          /
            \   \  |   | |   | |   |  __     /    /      
             \     |   | |   | |   | |   |  /    /   /
      \   \   \    |   | |   | |   | |   |      /   /
       \   \       |   | |   | |   | |   |     /   /     /
   \    \   \      |   | |   | |   | |   |        / /   / /
    \    \         |   | |   | |   | |   |         /   / /
     \     ____    |   | |   | |   | |   |        /     /
      \   | |  |   |   | |   | |   | |   |             /
          |_|  |   |   |_|   |_|   |_|   |
          |    |   /                     |
           \    \  |  v     v    v     v |
            \      Y                     |
             \        \                  |
              \         '                |  
               \          '             |
                 |         '            |
                  \                     |  
                    \                  |
                     |                 | 
                     |                | 
                
                
                """)
    exit(0)

def category():
    dct = {}
    review = {}
    cat = input("""Which category would you like to study?
    - food
    - animals
    - places 
    - professions
    - nouns_adj
    - phrases 
    - colors 
    - numbers

You choose: """)
    if 'food' in cat:
        dct = food
    elif 'animal' in cat:
        dct = animals
    elif 'place' in cat:
        dct = places
    elif 'profes' in cat:
        dct = professions
    elif 'noun' in cat:
        dct = noun_adj
    elif 'phrase' in cat:
        dct = phrases
    elif 'color' in cat:
        dct = colors
    elif 'numb' in cat:
        dct = numbers
    elif 'revi' in cat:
        dct = review
    else:
        print("""
        I'm not sure which category you want. Try again.""")
        category()
        
    correct = 0 
    incorrect = 0
    while len(dct) > 0:
        keys = list(dct.keys())
        k = random.choice(keys)
        v = dct[k]
        new = {k:v}
        print(f"""
_________________________________________________________________________

English word: {k}""")
        answer = input("""
        Type the Turkish translation and press <Enter> to check if you are correct.
        
Turkish word: """)
        answer = str.lower(answer)
        if answer == dct[k]:
            correct += 1
            print(f"""
        ***That's right! Try the next one!
        Score: Correct = {correct}, Incorrect = {incorrect}
_________________________________________________________________________""")
            review.update(new)           
            del dct[k]
        elif '0' in answer:
            print(f"""
_________________________________________________________________________

We can continue at a different time.
_____________________________
Here is what you got correct:
    {review}
_____________________________
The rest of the pairs are:
    {dct}
_____________________________""")
            success()
        elif answer != dct[k]:
            incorrect += 1
            print(f"""
            Mm, we'll try that one again later. 
            Score: Correct = {correct}, Incorrect = {incorrect}
            
            The answer should have been:
            ***{v}.
_________________________________________________________________________""")

            continue
        else:
            print("""There's been an error.""")
    print(f"""
_________________________________________________________________________
You've finished the lesson!!
    
Here is what you got correct:

    {review}
    """)
    success()
```
